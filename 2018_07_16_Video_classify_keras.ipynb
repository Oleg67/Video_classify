```python
def seperate_video_to_image(file_video, folder_name, period_image=1, verbose=True):
    """
    """
    import cv2
    video_count = 1
    image_count = 0
    image_freq = period_image # period chosen the image

    video = cv2.VideoCapture(file_video)

    while video.isOpened():
        success, image = video.read()
    
        if not np.all(image <= 0.01):
            if not os.path.exists(folder_name +'/video_' + str(video_count)):
                os.makedirs(folder_name +'/video_' + str(video_count))
    
        if np.any(image != None):
            image = cv2.resize(image, (224,224), interpolation=cv2.INTER_AREA)
        if not success:
            print 'Break'
            #break
            return 
        # seperate by black   
        if np.all(image <= 0.01):
            if image_count != 0:
                video_count += 1
            image_count = 0
        else:
            if (image_count % image_freq == 0):
                filename = folder_name +'/video_' +str(video_count).zfill(3) + '/' + str(image_count+1).zfill(5) +'_image.jpg'       
                
                if verbose :
                    print filename, '  ',image.shape

                success_write = cv2.imwrite(filename, image)
                if not success_write:
                    print 'No write'
                #    break
                    return
            image_count +=1
    
    print 'Done'
    
    print 'videos ', video_count
    numbers_of_images = np.zeros(video_count)

    for i,f in enumerate(os.listdir(folder_name)):
        numbers_of_images[i] = len(os.listdir(folder_name +'/' + f))
    
    print 'max_image_len ', int(numbers_of_images.max())
    print 'min_image_len ', int(numbers_of_images.min())
    return 
```

```python
def randomCreateListofImages(folder, max_length=50, n_folders=10):
    
    
    f_list = np.sort(os.listdir(os.path.join(os.getcwd(), folder)))
    f_list = np.vectorize(lambda x: os.path.join(os.getcwd(), folder, x))(f_list)
        
    return np.sort(np.random.choice(f_list, size=(n_folders,max_length)),axis=1)
```

```python
def CreateListofImages(folder, max_length=76):
    
    
    f_list = np.sort(os.listdir(os.path.join(os.getcwd(), folder)))
    f_list = np.vectorize(lambda x: os.path.join(os.getcwd(), folder, x))(f_list)
    
    f_lists = []
    f_size = len(f_list)
    if f_size  < max_length:
        return f_lists.append(f_list)
    j = 0
    for i in range(f_size - max_length):
        f_lists.append(f_list[i:max_length+i])
        
    return f_lists
```

```python
from scipy.misc import imread, imresize

def CreateTensorFromListofImages(image_list):
    
    
    X = np.full((len(image_list), 224, 224, 3), 0)
    #image_list = np.sort(np.random.choice(os.listdir(folder), size=max_length))
    for i,image in enumerate(image_list):
        x = imread(image)
        X[i] = x
    return X
```
